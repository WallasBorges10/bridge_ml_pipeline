```python
# Diagnóstico Rápido da Qualidade dos Dados
import pandas as pd

df = pd.read_csv("https://www.fhwa.dot.gov/bridge/nbi/2025/delimited/NY25.txt")
print("=== Dimensões ===")
print(df.shape)
print("\n=== Tipos de Dados ===")
print(df.dtypes.value_counts())
print("\n=== % de Valores Ausentes (Top 10) ===")
print((df.isnull().sum() / len(df) * 100).sort_values(ascending=False).head(10))
print("\n=== Amostra dos Dados ===")
df.head()
```
